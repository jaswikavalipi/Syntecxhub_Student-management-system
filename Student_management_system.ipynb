```python
import json
import os


class Student:
    def __init__(self, student_id, name, grade):
        self.student_id = student_id
        self.name = name
        self.grade = grade

    def to_dict(self):
        """Convert student object to dictionary"""
        return {"id": self.student_id, "name": self.name, "grade": self.grade}


class StudentManager:
    def __init__(self, filename="students.json"):
        self.filename = filename
        self.students = []
        self.load_students()

    def load_students(self):
        """Load students from JSON file"""
        if os.path.exists(self.filename):
            with open(self.filename, "r") as file:
                try:
                    data = json.load(file)
                    self.students = [Student(**s) for s in data]
                except json.JSONDecodeError:
                    self.students = []
        else:
            self.students = []

    def save_students(self):
        """Save students to JSON file"""
        with open(self.filename, "w") as file:
            json.dump([s.to_dict() for s in self.students], file, indent=4)

    def add_student(self, student_id, name, grade):
        """Add a new student (unique ID validation)"""
        if any(s.student_id == student_id for s in self.students):
            print("âŒ Error: Student ID must be unique!")
            return
        self.students.append(Student(student_id, name, grade))
        self.save_students()
        print("âœ… Student added successfully!")

    def update_student(self, student_id, name=None, grade=None):
        """Update student details"""
        for s in self.students:
            if s.student_id == student_id:
                if name:
                    s.name = name
                if grade:
                    s.grade = grade
                self.save_students()
                print("âœ… Student record updated!")
                return
        print("âŒ Student not found!")

    def delete_student(self, student_id):
        """Delete a student record"""
        for s in self.students:
            if s.student_id == student_id:
                self.students.remove(s)
                self.save_students()
                print("âœ… Student deleted successfully!")
                return
        print("âŒ Student not found!")

    def list_students(self):
        """Display all student records in formatted table"""
        if not self.students:
            print("âš ï¸ No students found.")
            return
        print("\nğŸ“‹ Student Records:")
        print("-" * 40)
        print(f"{'ID':<10} {'Name':<20} {'Grade':<10}")
        print("-" * 40)
        for s in self.students:
            print(f"{s.student_id:<10} {s.name:<20} {s.grade:<10}")
        print("-" * 40)


def main():
    manager = StudentManager()

    while True:
        print("\n=== ğŸ§‘â€ğŸ“ Student Management System ===")
        print("1. Add Student")
        print("2. Update Student")
        print("3. Delete Student")
        print("4. List Students")
        print("5. Exit")

        choice = input("Enter choice (1-5): ").strip()

        if choice == "1":
            student_id = input("Enter Student ID: ")
            name = input("Enter Name: ")
            grade = input("Enter Grade: ")
            manager.add_student(student_id, name, grade)

        elif choice == "2":
            student_id = input("Enter Student ID to update: ")
            name = input("Enter new Name (leave blank to skip): ")
            grade = input("Enter new Grade (leave blank to skip): ")
            manager.update_student(student_id, name or None, grade or None)

        elif choice == "3":
            student_id = input("Enter Student ID to delete: ")
            manager.delete_student(student_id)

        elif choice == "4":
            manager.list_students()

        elif choice == "5":
            print("ğŸ‘‹ Exiting... Goodbye!")
            break

        else:
            print("âŒ Invalid choice! Please try again.")


if __name__ == "__main__":
    main()
```
